__Single-layer Perceptron:__ This is the simplest feedforward neural Network and does not contain any hidden layer, Which means it only consists of a single layer of output nodes. This is said to be single because when we count the layers we do not include the input layer, the reason for that is because at the input layer no computations is done, the inputs are fed directly to the outputs via a series of weights.



```python
import random
from random import choice
mylist = ["perception", "planning", "control"]
print(random.choices(mylist, weights = [9, 1, 1], k = 14))
```

<center>Activation Function: Binary Step Function</center>



```python
activation = lambda x: 0 if x < 0 else 1
```

<center>Creating training data in form of list (train_x,train_y)</center>

```python
training_data = [ (np.array([0,0,1]), 0),
                 (np.array([0,1,1]), 0),
                 (np.array([1,0,1]), 0),
                 (np.array([1,1,1]), 1),
                ]
```

<center>Model Parameters</center>

```python
learning_rate = 0.2 
training_steps = 100 # number of epochs
```

<center> Complete Training </center>

We train our model in 100 training steps and using online or stochastic learning, which means that we use a single training example to make parameter update. Here the single training example is picked randomly

```python
# initialize weights 
W = np.random.rand(3)
print(W)
for i in range(training_steps):
    
    x, y = choice(training_data)
    
    l1 = np.dot(W, x)
    y_pred = activation(l1)
    
    error = y - y_pred
    update = learning_rate * error * x 
    W += update
#     print(W)
```

<center> Model Testing Output  </center>

```python
print("Predictions after training")
for x, _ in training_data:
    y_pred = np.dot(x, W)
    print("{}: {}".format(x[:2], activation(y_pred)))
```
